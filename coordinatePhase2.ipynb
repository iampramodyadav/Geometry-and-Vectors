# Coordinate lib file creation

```python
import re
import os

File1 = "Interface location and Areas.txt"                 #File1: Area, perimeter and centroid
File2 = "D4200614_Interface location and Orientation.txt"  #File2: Direction cosine reading
```

## File1: Area, perimeter and centroid

```python
# -----------------------File1: Area, perimeter and centroid-----------------------
with open(File1) as f1:
    lines1 = f1.readlines()
```

```python
PartName = []
n = len(lines1)
for line in lines1:
    text = line
    pattern = r'Face of (\w+\s*\w+\s*\w+)\n*'
    match = re.search(pattern, text)
    if match:
        result = match.group(1)
        #         print(result)
        PartName.append(result)
        
PartName = [x.upper() for x in PartName]

print(PartName)
len(PartName)
```

```python
AREA = []
n = len(lines1)
for i in lines1:
    text = i
    pattern = r'Area\s\[mmÂ²\]\t*(\d+\.\d+[E-]*\d*)\n*'
    match = re.search(pattern, text)
    if match:
        result = match.group(1)
        #         print(result)
        AREA.append(result)

AREA = [float(x) / 1000000 for x in AREA]
AREA = [str(x) for x in AREA]
print(AREA)
len(AREA)
```

```python
Peri = []
n = len(lines1)
for i in lines1:
    text = i
    pattern = r'Perimeter\s\[mm\]\t*(\d+\.\d+[E-]*\d*)\n*'
    match = re.search(pattern, text)
    if match:
        result = match.group(1)
        #         print(result)
        Peri.append(result)

Peri = [float(x) / 1000 for x in Peri]
Peri = [str(x) for x in Peri]
print(Peri)
len(Peri)
```

```python
centroid = []
rad = []
n = len(lines1)

for ii in range(n):    
    text = lines1[ii]
    pattern = r'Centroid\s*\[mm\]\t*\s*(-?\d+\.\d+[E-]*\d*)\t*\s*(-?\d+\.\d+[E-]*\d*)\t*\s*(-?\d+\.\d+[E-]*\d*)\n*'
    match = re.search(pattern, text)
    
    if match:
        result1 = match.group(1)
        result2 = match.group(2)
        result3 = match.group(3)
        centroid.append([result1, result2, result3])
        try:
            textRad=lines1[ii+1]
            pattern2 = r'Curvature\s*\[mm\]\t*\s*(-?\d+\.\d+[E-]*\d*)\n*'
            match2 = re.search(pattern2, textRad)
            if match2:
                result1 = match2.group(1)
                rad.append(result1)      
            else:
                rad.append(0)
        except IndexError:
            rad.append(0)
            
centroid = [[float(y) / 1000 for y in x] for x in centroid]
centroid = [[str(y) for y in x] for x in centroid]
print(centroid)
print(len(centroid))

rad = [float(x) / 1000 for x in rad]
rad = [str(x) for x in rad]
print(rad)
print(len(rad))
```

## File2: Direction cosine reading

```python
# -----------------------File2: Direction cosine reading-----------------------
with open(File2) as f2:
    lines2 = f2.readlines()
```

```python
PartNameDC = []
NumListDC = []
n = len(lines2)

for line in lines2:
    text = line
    pattern = r'D\d+\s(\w\D\w+\s*\w+\s*\w+)\n*'
    match = re.search(pattern, text)
    if match:
        result = match.group(1)
        PartNameDC.append(result)

        pattern1 = r'[\s\t](-?\d+\.*\d*[E-]*\d*)'
        match1 = re.findall(pattern1, text)
        if match:
            result1 = match1
            NumListDC.append(result1)
            
PartNameDC = [x.upper() for x in PartNameDC]
print(PartNameDC)
print(len(NumListDC))
NumListDC = [[float(y) for y in x] for x in NumListDC]
NumListDC = [[str(y) for y in x] for x in NumListDC]

print(NumListDC)
print(len(NumListDC))
```
